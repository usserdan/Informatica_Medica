## Identificacin de missing values en Python

En la librara Pandas de Python, los valores perdidos se representan con $None$ y $NaN$ (acrnimo de _Not a Number_). Pandas asigna automticamente $NaN$ si el valor de una celda es un string vaco '', NA o NaN.


Sin embargo, hay ocasiones en las que los valores perdidos se representan con un valor diferente a los antes mencionados, por ejemplo, con un $0$ en el caso de una variable numrica o $?$, como en el caso de nuestro conjunto de datos. En este ltimo caso, lo que debemos hacer es reemplazar el valor de estos valores perdidos representados con $?$ por el valor $NaN$, y para ello Pandas nos permite pasar una lista de valores que queremos considerar en todas las columnas como valores perdidos mediante el parmetro na_values:

```python
na_values = ["?"]

# Cargamos el dataset
df2 = pd.read_excel('dataset_imputacion.xlsx', na_values = na_values)

# Aadimos el nombre de cada variable
df2.columns = list_
df2.head()
```

## $Sklearn.impute$
Este paquete proporciona 2 tipos de algoritmos de imputacin para completar los valores faltantes:

### 1. $SimpleImputer$

Se utiliza para imputaciones en <b>conjuntos de datos univariados</b>. $SimpleImputer$ permite imputar valores en cualquier columna de caractersticas utilizando solo los valores faltantes en ese espacio de caractersticas.

1. Se proporcionan diferentes estrategias para imputar datos, como imputar con:<br>
    * un valor constante o <br>
    * usar mtodos estadsticos como:<br>
        - la media, <br>
        - la mediana o <br>
        - la moda<br>
<br>
2. Para las representaciones de datos categricos, tiene soporte para la estrategia 'ms frecuente' , que es como la moda de los valores numricos .

```python
from sklearn.impute import SimpleImputer
```

### Imputar la edad usando la media ($mean$) de las edades

```python
# La media de las edades en el dataset
df3.Paciente_Edad.mean()
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
X = df3[['Paciente_Edad']]
X = imp.fit_transform(X)
X
```

### Imputar la edad usando la mediana ($median$) de las edades

```python
df3.Paciente_Edad.median()
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='median')
X = df3[['Paciente_Edad']]
X = imp.fit_transform(X)
X
```

### Imputar la edad usando un valor constante  ($constant$)

```python
imp = SimpleImputer(missing_values=np.nan, strategy='constant' , fill_value=666)
X = df3[['Paciente_Edad']]
X = imp.fit_transform(X)
X
```

### Imputar el grupo etario usando la moda ($most\_frequent$) del Grupo Edad

```python
df3.Grupo_Edad
```

```python
df3.Grupo_Edad.mode()
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
X = df3[['Grupo_Edad']]
X = imp.fit_transform(X)
X
```

```python
pd.DataFrame(X)
```

### 2. $IterativeImputer$

Se utiliza para imputaciones en <b>conjuntos de datos multivariados</b>, los <b>conjuntos de datos</b> multivariados son conjuntos de datos que tienen ms de dos variables o columnas de caractersticas por observacin . $IterativeImputer$ nos permite hacer uso de todo el conjunto de datos de las columnas de caractersticas disponibles para imputar los valores faltantes.

```python
# Se extraen solo las variables numricas.

varNum = ['Paciente_Edad','Urg_Demora1_Consulta_(Minutos)','Urg_Demora_Triaje_(Minutos)','Urg_Demora2_Consulta_(Minutos)']
varNum
```

```python
# Requiere explicitamente es funcin experimental
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
imp = IterativeImputer(max_iter=10 , random_state=0)
X = df3[varNum]
X = imp.fit_transform(X)
df5 = pd.DataFrame(X , columns=varNum)
df5
```

### 3. $KNNImputer$

Es una tcnica de imputacin de <b>datos multivariantes</b> que se utiliza para completar los valores faltantes mediante el <b>enfoque K-Neighbors Neighbors</b>. Cada valor faltante se completa con el <b>valor medio</b> de los <b>$n$ vecinos ms cercanos</b> encontrados en el conjunto de entrenamiento, ponderados o no ponderados.

```python
# Se hace la importacin de la librera
from sklearn.impute import KNNImputer
```

```python
imp = KNNImputer(n_neighbors=3 , weights='uniform')
X = df[varNum]
X = imp.fit_transform(X)
df6 = pd.DataFrame(X , columns=varNum)
df6
```
