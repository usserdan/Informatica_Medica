### Ejercicio prctica

Construyendo una Red Neuronal Recurrente LSTM, hacer un sistema que haga la prediccin del comportamiento de las accionens de Apple entre Enero del 2006 y Diciembre del 2017. Se tomarn 60 datos del comportamiento de las acciones y se har la prediccin del dato 61. Para este ejercicio solo trabajaremos con el valor ms alto de la accin.

Tomado de: www.codificandobits.com

```python
import numpy as np
np.random.seed(4)
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
def graficar_predicciones(real, prediccion):
    plt.plot(real[0:len(prediccion)],color='red', label='Valor real de la accin')
    plt.plot(prediccion, color='blue', label='Prediccin de la accin')
    plt.ylim(1.1 * np.min(prediccion)/2, 1.1 * np.max(prediccion))
    plt.xlabel('Tiempo')
    plt.ylabel('Valor de la accin')
    plt.legend()
    plt.show()
```

# Lectura de los datos

```python
# https://github.com/angelower/infomed20251/blob/main/data/AAPL_2006-01-01_to_2018-01-01.csv
!wget https://raw.githubusercontent.com/angelower/infomed20251/main/data/AAPL_2006-01-01_to_2018-01-01.csv
```

```python
dataset = pd.read_csv('AAPL_2006-01-01_to_2018-01-01.csv'
                      , sep=';'
                      , index_col='Date'
                      , parse_dates=['Date']
                        )
dataset.head()
```

```python
# Sets de entrenamiento y validacin
# La LSTM se entrenar con datos de 2016 hacia atrs. La validacin se har con datos de 2017 en adelante.
# En ambos casos slo se usar el valor ms alto de la accin para cada da
set_entrenamiento = dataset[:'2016'].iloc[:,1:2]
set_validacion = dataset['2017':].iloc[:,1:2]

set_entrenamiento['High'].plot(legend=True)
set_validacion['High'].plot(legend=True)
plt.legend(['Entrenamiento (2006-2016)', 'Validacin (2017)'])
plt.show()
```

```python
# Normalizacin del set de entrenamiento
sc = MinMaxScaler(feature_range=(0,1))
set_entrenamiento_escalado = sc.fit_transform(set_entrenamiento)
```

```python
# La red LSTM tendr como entrada "time_step" datos consecutivos, y como salida 1 dato (la prediccin a
# partir de esos "time_step" datos). Se conformar de esta forma el set de entrenamiento
time_step = 60
X_train = []
Y_train = []
m = len(set_entrenamiento_escalado)

for i in range(time_step,m):
    # X: bloques de "time_step" datos: 0-time_step, 1-time_step+1, 2-time_step+2, etc
    X_train.append(set_entrenamiento_escalado[i-time_step:i,0])

    # Y: el siguiente dato
    Y_train.append(set_entrenamiento_escalado[i,0])
X_train, Y_train = np.array(X_train), np.array(Y_train)

# Reshape X_train para que se ajuste al modelo en Keras
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

# Red LSTM

```python
dim_entrada = (X_train.shape[1],1)
dim_salida = 1
na = 50

modelo = Sequential()
modelo.add(LSTM(units=na, input_shape=dim_entrada))
modelo.add(Dense(units=dim_salida))
modelo.compile(optimizer='rmsprop', loss='mse')
modelo.fit(X_train,Y_train,epochs=20,batch_size=32)
```


# Validacin (prediccin del valor de las acciones)


```python
x_test = set_validacion.values
x_test = sc.transform(x_test)

X_test = []
for i in range(time_step,len(x_test)):
    X_test.append(x_test[i-time_step:i,0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))

prediccion = modelo.predict(X_test)
prediccion = sc.inverse_transform(prediccion)
```

# Graficar resultados

```python
graficar_predicciones(set_validacion.values,prediccion)
```
